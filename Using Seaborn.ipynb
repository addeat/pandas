```python
titanic = sns.load_dataset('titanic')

print(titanic.head())
print()
print(titanic.info())
```

```python
sns.set_style('darkgrid')
```

```python
fig = plt.figure(figsize = (15, 5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.regplot(x= 'age', y='fare' , data = titanic, ax = ax1)

sns.regplot(x= 'age', y='fare' , data = titanic, ax = ax2, fit_reg = False, color = 'orange')
plt.show()
```

```python
fig = plt.figure(figsize = (15, 5))

ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)

sns.distplot(titanic['fare'], ax = ax1)
sns.kdeplot(x = 'fare', data = titanic , ax = ax2 )
sns.histplot(x = 'fare', data = titanic, ax = ax3)

ax1.set_title('titanic fare = hist/ked')
ax1.set_title('titanic fare = ked')
ax1.set_title('titanic fare = hist')
```

```python
table = titanic.pivot_table(index = ['sex'], columns = ['class'], aggfunc = 'size')

sns.heatmap(table,                    # 
           annot = True, fmt = 'd',   #    ,  
           cmap = 'YlGnBu',           # 
           linewidth = .5,            #  
           cbar = False)              #  
```

```python
fig = plt.figure(figsize = (15, 5))

ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

sns.stripplot(x = 'class', y = 'age', data = titanic, hue = 'sex', ax = ax1)
sns.swarmplot(x = 'class', y = 'age', data = titanic, hue = 'sex',ax = ax2)

ax1.set_title('Strip Plot')
ax2.set_title('Swarm Plot')

plt.show()
```

```python
fig = plt.figure(figsize = (15, 5))

ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)

sns.barplot(x = 'sex', y = 'survived', data = titanic, ax = ax1)

sns.barplot(x = 'sex', y = 'survived', hue = 'class', data = titanic, ax = ax2)

sns.barplot(x = 'sex', y = 'survived', hue = 'class', dodge = False , data = titanic, ax = ax3)

```

```python
fig = plt.figure(figsize = (15, 5))

ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)

sns.countplot(x = 'class', palette = 'Set2', hue = 'survived', data = titanic, ax = ax1)
sns.countplot(x = 'class', hue = 'who', palette = 'Set1', data = titanic, ax = ax2)
sns.countplot(x = 'class', hue = 'who', palette = 'Set3', dodge = False, data = titanic, ax = ax3)

plt.show()
```

```python
fig = plt.figure(figsize = (15, 10))
ax1 = fig.add_subplot[[2, 2, 1]]
ax2 = fig.add_subplot[[2, 2, 2]]
ax3 = fig.add_subplot[[2, 2, 3]]
ax4 = fig.add_subplot[[2, 2, 4]]
sns.boxplot (x = 'alive',y = 'age', data = titanic, ax = a1)
sns.boxplot (x = 'alive',y = 'age', hue = 'sex', data = titanic, ax = a2)
sns.boxplot (x = 'alive',y = 'age', data = titanic, ax = a3)
```
