```python
# 데이터 갯수를 확인해주는 메소드
df.shape
```

```python
# 데이터에 관한 기본 정보들을 제공해주는 메소드
df.info()
```

```python
# 산술 데이터 (평균, 표준편차, 최댓값, 최솟값, 중간값 등등..)를 요약 출력
print(df.describe())
print(df.describe(include = 'all'))
```

```python
# 열데이터 갯수 정보 출력 series type
df.count()
```

```python
#value_counts()메소드는 고유값을 세는데 사용.   이 파일에선 origin의 고유값을 세볼것임. 1은 미국 2는 유럽 3은 일본이라함.
unique_values = df['origin'].value_counts()
print(unique_values)
```

### 통계 함수

```python
# 평균값
# 전체 평균값
df_mean = df.mean()
print(df_mean)       # -> 여기서 의미 있을 것 같은 변수를 추출하는 프로그램은 어떻게만들지
print()
# 특정 열 평균 값
df_mw_mean = df[['mpg', 'weight']].mean()
print(df_mw_mean)
```

```python
# 중간값
df_median = df.median()
print(df_median)
print()
print(type(df_median))
```

```python
# 최댓값
df_max = df.max()
print(df_max)
print()

# 일부 최댓값 출력
df_mw_max = df[['mpg']].max()
print(df_mw_max)
print(type(df_mw_max))
```

```python
# 표준 편차

df_std = df.std()
print(df_std)
print()

# 특정 열 표편
df_mw_std = df[['mpg']].std()
print(df_mw_std)
```

```python
#상관계수 (correlation)
df_corr = df.corr()
print(df_corr)
print()

# 두 열 사이의 상관괸계 

df_mw_corr = df[['mpg', 'displacement']].corr()
print(df_mw_corr)
```
