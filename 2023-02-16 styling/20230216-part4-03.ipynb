```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic=sns.load_dataset('titanic')
titanic.head(10)
```

```python
titanic.info()
```

```python
sns.set_style('darkgrid')
```

```python
fig=plt.figure(figsize=(9,3))
ax1=fig.add_subplot(1,2,1)
ax2=fig.add_subplot(1,2,2)

sns.regplot(x='age',y='fare',
           data=titanic,
           ax=ax1)

sns.regplot(x='age',y='fare',
           data=titanic,
           ax=ax2,
           fit_reg=False)

plt.show()
```

```python
fig=plt.figure(figsize=(15,5))

ax1=fig.add_subplot(1,3,1)
ax2=fig.add_subplot(1,3,2)
ax3=fig.add_subplot(1,3,3)

sns.distplot(titanic['fare'],ax=ax1)

sns.kdeplot(x='fare',data=titanic,ax=ax2)
#sns.distplot(titanic['fare'],ax=ax2,hist=False)

sns.histplot(x='fare',data=titanic,ax=ax3)
#sns.distplot(titanic['fare'],ax=ax3,kde=False)

ax1.set_title('titanic fare - hist/kde')
ax2.set_title('titanic fare - kde')
ax3.set_title('titanic fare - hist')


plt.show()
```

```python
table=titanic.pivot_table(index=['sex'],columns=['class'],aggfunc='size')
```

```python
table
```

```python
sns.heatmap(table, annot=True, fmt='d',
           cmap='YlGnBu', linewidth=0.5 , cbar=False)


plt.show()
```

```python
fig=plt.figure(figsize=(15,5))
ax1=fig.add_subplot(1,2,1)
ax2=fig.add_subplot(1,2,2)

sns.stripplot(x='class',y='age',data=titanic,hue='sex',ax=ax1)
sns.swarmplot(x='class',y='age',data=titanic,hue='sex',ax=ax2)

plt.show()
```

```python
fig=plt.figure(figsize=(15,5))
ax1=fig.add_subplot(1,3,1)
ax2=fig.add_subplot(1,3,2)
ax3=fig.add_subplot(1,3,3)

sns.barplot(x='sex',y='survived',data=titanic,ax=ax1)

sns.barplot(x='sex',y='survived',hue='class',data=titanic,ax=ax2)

sns.barplot(x='sex',y='survived',hue='class',dodge=False,data=titanic,ax=ax3)

```

```python
fig=plt.figure(figsize=(12,3))
ax1=fig.add_subplot(1,3,1)
ax2=fig.add_subplot(1,3,2)
ax3=fig.add_subplot(1,3,3)

sns.countplot(x='class',palette='Set1',data=titanic,ax=ax1)

sns.countplot(x='class',palette='Set2',data=titanic,hue='who',ax=ax2)

sns.countplot(x='class',palette='Set3',data=titanic,dodge=False,hue='who',ax=ax3)


```

```python
fig=plt.figure(figsize=(8,4))
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)

sns.set_style('white')

sns.boxplot(x='alive',y='age',data=titanic,ax=ax1)
sns.boxplot(x='alive',y='age',data=titanic,hue='sex',ax=ax2)

sns.violinplot(x='alive',y='age',data=titanic,ax=ax3)
sns.violinplot(x='alive',y='age',hue='sex',data=titanic,ax=ax4)

plt.show()
```

```python
sns.set_style('whitegrid')

j1=sns.jointplot(x='fare',y='age',data=titanic,height=3)
j2=sns.jointplot(x='fare',y='age',kind='reg',data=titanic,height=3)
j3=sns.jointplot(x='fare',y='age',kind='hex',data=titanic,height=3)
j4=sns.jointplot(x='fare',y='age',kind='kde',data=titanic,height=3)


plt.show()
```
