```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from matplotlib import font_manager, rc
font_path = "./malgun.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
df=pd.read_csv('./auto-mpg.csv',header=None)
```

```python
df.columns = ['mpg','cylinders','displacement','horsepower','weight',
              'acceleration','model year','origin','name']
```

```python
plt.style.use('classic')

df['mpg'].plot(kind='hist',bins=10,color='coral',figsize=(5,2))

plt.title('mpg histogram')
plt.xlabel('mpg')

plt.show()
```

```python
df['mpg'].describe()
```

```python
df.plot(kind='scatter',x='weight',y='mpg',c='red',s=7,figsize=(5,3))

plt.title('scatter plot- mpg vs weight')

plt.show()
```

```python
cylinders_size=df.cylinders/df.cylinders.max()*300
```

```python
df.plot(kind='scatter',x='weight',y='mpg',marker='+',cmap='plasma',
        c=cylinders_size,figsize=(10,5),alpha=0.3)

plt.title('scatter plot- mpg vs weight')

plt.savefig('./sample01.png')
plt.savefig('./sample02.png',transparent=True)


plt.show()
```

```python
df_origin=df.groupby('origin').sum()
```

```python
df_origin.index=['USA','EU','JPN']
```

```python
df_origin
```

```python
plt.style.use('default')

df_origin['count'].plot(kind='pie',figsize=(5,3),autopct='%1.1f%%',startangle=10,
                       colors=['chocolate','bisque','cadetblue'])

plt.title('Model Origin',size=15)
plt.axis('equal')

plt.legend(labels=df_origin.index,loc='upper right')

plt.show()
```
